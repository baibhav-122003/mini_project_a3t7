# Setup Catalog, Schema, and Volumes

```python
%sql
CREATE CATALOG IF NOT EXISTS azure3_team7_project;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS azure3_team7_project.bronze
COMMENT 'Raw ingestion layer - immutable, append-only, exact copy of source data';

CREATE SCHEMA IF NOT EXISTS azure3_team7_project.silver
COMMENT 'Cleaned, validated, deduplicated and typed data';

CREATE SCHEMA IF NOT EXISTS azure3_team7_project.gold
COMMENT 'Business-aggregated, consumption-ready data';

CREATE SCHEMA IF NOT EXISTS azure3_team7_project.platform
COMMENT 'Pipeline audit logs and metadata - not business data';
```

```python
%sql
SHOW SCHEMAS IN azure3_team7_project;
```

```python
%sql
CREATE VOLUME IF NOT EXISTS azure3_team7_project.bronze.landing
COMMENT 'Raw file landing zone - files pushed by ERP, WMS, Kafka, APIs';

CREATE VOLUME IF NOT EXISTS azure3_team7_project.bronze.checkpoints
COMMENT 'Streaming checkpoints for bronze ingestion jobs';

CREATE VOLUME IF NOT EXISTS azure3_team7_project.silver.checkpoints
COMMENT 'Streaming checkpoints for silver transformation jobs';

CREATE VOLUME IF NOT EXISTS azure3_team7_project.gold.checkpoints
COMMENT 'Checkpoints for gold aggregation jobs using trigger availableNow';
```
