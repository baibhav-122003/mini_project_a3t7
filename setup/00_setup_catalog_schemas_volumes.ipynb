```python
# Create Catalogs
# Change Managed Location according to the current ADLS resource. I got the root path from the default catalog which came with the workspace.

MANAGED_LOCATION = "abfss://unity-catalog-storage@dbstorage6qpvsjlux3tpm.dfs.core.windows.net/7405605756557198"

spark.sql(f"""
        CREATE CATALOG IF NOT EXISTS azure3_team7_project
        MANAGED LOCATION '{MANAGED_LOCATION}';
    """)
```

```python
# Create Schemas

schemas = [
    ("bronze",   "Raw ingestion layer - immutable, append-only, exact copy of source data"),
    ("silver",   "Cleaned, validated, deduplicated and typed data"),
    ("gold",     "Business-aggregated, consumption-ready data"),
    ("platform", "Pipeline audit logs and metadata - not business data"),
]

for schema, comment in schemas:
    spark.sql(f"""
        CREATE SCHEMA IF NOT EXISTS azure3_team7_project.{schema}
        COMMENT '{comment}'
    """)
    print(f"[CREATED] azure3_team7_project.{schema}")

spark.sql("SHOW SCHEMAS IN azure3_team7_project").show()
```

```python
# Create Volumes

volumes = [
    ("bronze", "landing",     "Raw file landing zone - files pushed by ERP, WMS, Kafka, APIs"),
    ("bronze", "checkpoints", "Streaming checkpoints for bronze ingestion jobs"),
    ("silver", "checkpoints", "Streaming checkpoints for silver transformation jobs"),
    ("gold",   "checkpoints", "Checkpoints for gold aggregation jobs using trigger availableNow"),
]

for schema, volume, comment in volumes:
    spark.sql(f"""
        CREATE VOLUME IF NOT EXISTS azure3_team7_project.{schema}.{volume}
        COMMENT '{comment}'
    """)
    print(f"[CREATED] azure3_team7_project.{schema}.{volume}")

# List volumes
for schema in ["bronze", "silver", "gold"]:
    print(f"\nVolumes in {schema}:")
    spark.sql(f"SHOW VOLUMES IN azure3_team7_project.{schema}").show()
```
