# Setup Audit Tables for Lineage Tracking and Tracability

```python
%sql
CREATE TABLE IF NOT EXISTS azure3_team7_project.platform.pipeline_audit (
  run_id            STRING        NOT NULL,
  pipeline_name     STRING        NOT NULL,
  source_table      STRING,
  target_table      STRING        NOT NULL,
  start_time        TIMESTAMP     NOT NULL,
  end_time          TIMESTAMP,
  rows_read         BIGINT,
  rows_written      BIGINT,
  rows_rejected     BIGINT,
  status            STRING        NOT NULL,  -- RUNNING, SUCCESS, FAILED
  error_message     STRING,
  spark_app_id      STRING,
  extra_metadata    MAP<STRING, STRING>
)
USING DELTA
TBLPROPERTIES (
  'delta.autoOptimize.optimizeWrite' = 'true',
  'delta.autoOptimize.autoCompact'   = 'true'
)
COMMENT 'Central audit log for all pipeline runs across bronze, silver and gold layers';
```

```python
%sql
DESCRIBE TABLE azure3_team7_project.platform.pipeline_audit;
```
